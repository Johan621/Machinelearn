```python
import pandas as pd 
#Read the Us Pumpkins csv file
pumpkins = pd.read_csv('../Messydata/US-pumpkins.csv')

# Filter out the any empty rows that don't use price per bushel
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel',case = True,regex = True)]
pumpkins.head() 
```

```python
#Count how many number of cells in each column are empty
pumpkins.isnull().sum()
```

```python
#Define the columns we want to keep
new_columns = ['Package','Month','Date','Low Price','High Price']
#Drop all other columns
pumpkins = pumpkins.drop([c for c in pumpkins if c not in new_columns],axis=1)
```

```python
#Calculate the average price from the high and low prices
price = (pumpkins['Low Price'] + pumpkins['High Price'])/2
#Get the month from the date column
month = pd.DatetimeIndex(pumpkins['Date']).month
#Create a new dataframe with these columns
new_pumpkins = pd.DataFrame({'Month':month,'Package':pumpkins['Package'],'Low Price':pumpkins['Low Price'],'High Price':pumpkins['High Price'],'Price':price})
```

```python
#Convert the price of all cells price by 1 1/9 bushels by dividing by 1 1/9
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'),'Price'] = price/(1 + 1/9)
#Convert the price of all cells price by 1/2 bushels by dividing by 1/2
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'),'Price'] = price/(1/2)

#print the first few rows from the data
new_pumpkins.head()
```
