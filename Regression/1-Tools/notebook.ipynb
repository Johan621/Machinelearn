```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model,model_selection

#Load the Diabetes data 
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])

#Extract the column of index 2
X = X[:, 2]
print(X.shape)

#Reshape to 2D-array
X = X.reshape((-1,1))
print(X.shape)
```

```python
#Split the data into training and testing data
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size=0.33)
```

```python
# Create a Linear Regression model and train with our data
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
#predict/test using our test data
y_prec = model.predict(X_test)
```

Now we are going to test/predict our prediction using matplotlib Graph plots 

```python
#Create a scatter plots 
plt.scatter(X_test,y_test,color = 'black')
#Plot the prediction line
plt.plot(X_test,y_prec,color = 'Blue',linewidth = 3)

#Add labels and Title to the Graph
plt.xlabel("Scaled BMI's")
plt.ylabel("Disease Progression")
plt.title("A Graph plot showing Diabetes progression against BMI")

#Draw the plot
plt.show()
```
